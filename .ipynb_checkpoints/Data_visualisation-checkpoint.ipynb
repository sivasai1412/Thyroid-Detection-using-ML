```python
y = df['binaryClass']
sns.countplot(y)
target_temp = df.value_counts()
```

```python
#correlation heatmap
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap='terrain', linewidths=0.1)
fig=plt.gcf()
fig.set_size_inches(8, 6)
plt.show()
```

```python
#creating the different distplots

plt.figure(figsize=(12, 10))
plt.subplot(221)
sns.distplot(df[df['binaryClass'] == 0].age)
plt.title("Age of patients without thyroid disease")
plt.subplot(222)
sns.distplot(df[df['binaryClass'] == 1].age)
plt.title("Age of patients with thyroid disease")
plt.subplot(223)
sns.distplot(df[df['binaryClass'] == 0].FTI)
plt.title("Max FTI value of patients without thyroid disease")
plt.subplot(224)
sns.distplot(df[df['binaryClass'] == 1].FTI)
plt.title("Max FTI value of patients with thyroid disease")
plt.show()
```

```python
plt.figure(figsize=(13, 6))
plt.subplot(121)
sns.violinplot(x='binaryClass', y='FTI', data=df, inner=None)
sns.swarmplot(x='binaryClass', y='FTI', data=df, color='w', alpha=0.5)

plt.subplot(122)
sns.swarmplot(x='binaryClass', y='FTI', data=df)
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.subplot(131)
sns.pointplot(x='sex', y='binaryClass', data=df)
plt.legend(['male: 0', 'female: 1'])
plt.subplot(132)
sns.barplot(x='sick', y='binaryClass', data=df)
plt.legend(['yes = 1', 'no = 0'])
plt.subplot(133)
sns.countplot(x = 'T3', hue='binaryClass', data=df)
plt.show()
```
